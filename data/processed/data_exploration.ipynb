## Subscriptions Dataset

Rows: 600  
Columns: 17

Each row represents a customer and their plan and billing cycle. Along with that, it has additional information such as signup date, churned date, company size, and more.

287 missing values in churn date and churn reason due to companies not leaving

**Billing cycle:**
* Monthly
* Annual

**Plans:**
* starter
* professional
* business
* enterprise

**support_tickets_12mo:** How many support requests were submitted during the last 12 months  

**nps_score (NPS):** How likely are you to recommend us?

Dataset inspection revealed no duplicate records and no significant data quality issues. Missing values in churn-related fields were expected for active customers.
